```python
import pandas as pd
import numpy as np
```

```python
report=pd.read_csv('file.csv')
```

```python
report
```

```python
maths=np.random.randint(0,100,size=10)
```

```python
maths
```

```python
s=pd.Series(maths,index=np.arange(0,10))
```

```python
report=report.assign(maths=s)
```

```python
report
```

```python
# sortin one column 
report.sort_values(by='maths')
```

```python
report.sort_values(by='maths',ascending=False)
```

```python
# sort descending by multiple columns
report.sort_values(['maths','Points'],ascending=False)# it creates a copy 
```

```python
# it overides the values 
report.sort_values(['maths','Points'],ascending=False,inplace=True)
```

```python
report
```

```python
report.sort_values ('Student Name',ascending =True,key=lambda col:col.str.lower())
```
