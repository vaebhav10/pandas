```python
import numpy as np
import pandas as pd 

lap=pd.read_csv("laptops.csv",encoding='latin1')
lap=lap.rename(columns={'Price (Euros)':'price'})# renaming the price column
lap['price']=lap['price'].str.replace(",",'.')
lap['price'] =pd.to_numeric(lap['price'])
del lap['Operating System Version']# delating a column
lap['Screen Size'] =lap['Screen Size'].str.replace('"','')
lap['Screen Size'] =pd.to_numeric(lap['Screen Size'])
```

```python
lap.head(2)
```

```python
# exercise 
# find the biggest and smallest screen size in laptops of each company using the sort_values() and duplicated ()
# methods (keep "first " ,"last " and false)
```

```python
partition=lap.sort_values(['Screen Size','Manufacturer'])
```

```python
partition.head()
```

```python
partition.drop_duplicates(['Manufacturer'],keep='first')[['Manufacturer','Screen Size']]# smallest screen 
```

```python
partition.drop_duplicates(['Manufacturer'],keep='last')[['Manufacturer','Screen Size']]# companies with biggest screen size 
```
