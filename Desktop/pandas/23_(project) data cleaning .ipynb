2 Identifying missing data 

```python
# number of rows missing in each column
netflix.isnull().sum().sort_values(ascending=False) # shows the number of NAN 
```

```python
netflix['director'].isnull().mean()
```

```python
netflix.columns
```

```python
# % of rows missing in each column
for column in netflix.columns:
    per=netflix[column].isnull().mean()
    print(column+': '+str(round(per*100,2)))
    # print(per)
```

3 Dealing with missing Data 

```python
# remove a column or row with  .drop ,dropna, isnull
# drop column 
netflix.drop('director',axis=1)
```

```python
# drop rows 
noval=netflix[netflix['director'].isnull()].index
```

```python
# deleting rows with NAN values
netflix.drop(noval,axis=0)
```

```python
#isnull 
# netflix=[netflix['director'].isnull()]
netflix=netflix.dropna(subset=['director'])
```

```python
# use fillna() to replace NAN by the mean , mean (numeric data ) or mode 
mode=''.join(netflix['rating'].mode())
```
