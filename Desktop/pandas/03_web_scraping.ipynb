```python
# Replace commas with periods in the Price column
lap['Price (Euros)'] = lap['Price (Euros)'].str.replace(',', '.')

# Convert the column to numeric type
lap['Price (Euros)'] = pd.to_numeric(lap['Price (Euros)'], errors='coerce')

# Now create the tier column
lap['tier'] = np.where(lap['Price (Euros)'] > 2000, 'expensive', 'cheap')
```
