```python
import numpy as np
```

```python
# creating a random numbers for our new score columns 
score1 =np.random.randint (1,100,size =13)
score2 =np.random.randint (1,100,size =13)
```

```python
# creating a series using the random numbers 

series1 =pd.Series(score1,index=np.arange(0,13))
series2 =pd.Series(score2,index=np.arange(0,13))
```

```python
# using assign() to add multiple columns
report=report.assign(score1=series1,score2=series2)
```
