```python
import numpy as np
import pandas as pd 

lap=pd.read_csv("laptops.csv",encoding='latin1')
lap=lap.rename(columns={'Price (Euros)':'price'})# renaming the price column
lap['price']=lap['price'].str.replace(",",'.')
lap['price'] =pd.to_numeric(lap['price'])
del lap['Operating System Version']# delating a column
lap['Screen Size'] =lap['Screen Size'].str.replace('"','')
lap['Screen Size'] =pd.to_numeric(lap['Screen Size'])
```
