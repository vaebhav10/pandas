```python
import pandas as pd 
import numpy as np

df=pd.read_csv('file.csv')
del df['Unnamed: 0']
df=df.set_index('Student Name')
import random 
point1=[random.randint(150,180)for _ in range(13)]
df.rename(columns={'Points':'Marks'},inplace=True)
df['Height']=point1

df=df.dropna()# delete rows with any NaN values
```

1 Copy()

1.1 deep =True

```python
# deep = True by default ( Modifications to the data or indices of the copy )
df_copy=df.copy()
```

```python
# update value in original dataframe 
df.loc['Vaibhav Kumar','Height']=170
```

```python
df
```

```python
df_copy
```

1.2 deep =False

```python
# deep = False ( Any changes to the data of the original will be reflected )
df_s_copy=df.copy(deep=False)# its called shallow copy
```

```python
df.loc['Vaibhav Kumar','Height']=190
```

```python
df
```

```python
df_s_copy
```

1.3 Simple assignment

```python
# make a copy
df_new_copy=df# same as shallow copy 
```

```python
df.loc['Vaibhav Kumar','Height']=200
```
