1 Vertical concatenation

```python
import numpy as np 
import pandas as pd
```

```python
movies=pd.read_csv('IMDb movies.csv',low_memory =False)
ratings=pd.read_csv('IMDb ratings.csv')
```

```python
movies.columns
```

```python
movies=movies[['imdb_title_id', 'title', 'year','genre','country']] 
```

```python
ratings.columns
```

```python
ratings=ratings[['imdb_title_id','total_votes','mean_vote']]
```

```python
movies.head()
```

```python
ratings.head()
```

1 Exercise

```python
# extract a 50% sample of the original dataframe (movies)
movies['imdb_title_id'].frac=0.5
```

```python
h_movies=movies.sample(frac=0.5)
```

```python
h_movies.shape
movies.shape
```

```python
#concatenate both h_movies and ratings vertically
cat=pd.concat([movies,h_movies],axis=0)
```

```python
cat.shape
```

```python
cat
```

```python
cat.set_index(['imdb_title_id'])
```

```python
# concatenating movies and rating based on id
movies.merge(ratings,on='imdb_title_id',
             how='outer',indicator=True)
```
