```python
import pandas as pd 
import numpy as np

df=pd.read_csv('file.csv')
del df['Unnamed: 0']
df=df.set_index('Student Name')
import random 
point1=[random.randint(150,180)for _ in range(13)]
df.rename(columns={'Points':'Marks'},inplace=True)
df['Height']=point1

df=df.dropna()# delete rows with any NaN values
```

1 Lambda function 

```python
# basic function 
def sum_values(a,b):
    s=a+b
    return s
    
```

```python
sum_values(4,5)
```

```python
# lambda function
sum_lmd=lambda a,b:a+b# before : its input part and after that its output
```

```python
sum_lmd(4,5)
```

2 Apply + Lambda function

```python
# use lambda function to convert height seris to meters 

df["Height"].apply(lambda r:r/ 100)
# same as df['Height']/100
```

```python
df.head(3)
```

```python
# use lambda function to convert 'Exam Name' in upper case
df["Exam Name"].apply(lambda s: s.upper())

#alternative with str attribute
df['Exam Name'].str.upper()
```

```python
# use lambda function to get the year of the data
df['Date'] =df['Date'].astype('datetime64[ns]')
```

```python
df['Date'].apply(lambda a: a.year)
# df['Date'].dt.year
```

```python
# apply lambda function to dataframe in order to calculate sum of height and marks
k=lambda row: row['Marks']+row['Height']
df.apply(k,axis=1)
```
