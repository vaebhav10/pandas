```python
import numpy as np
import pandas as pd
```

```python
lap=pd.read_csv('laptops.csv',encoding ='latin1')
```

```python
lap.head(3)
```

```python
# renaming price column
lap=lap.rename(columns={'Price (Euros)':'price'})
```

```python
lap.head()
```

```python
lap['price']=lap['price'].str.replace(',','.')
```

```python
lap['price']= pd.to_numeric(lap['price'])
```

```python
lap.head(2)
```

```python
# creating a column based on price
np.where(lap['price']>1000,"expensive","Affordable")
```

```python
lap['status']=np.where(lap['price']>1000,"expensive","Affordable")
```

```python
lap.head(2)
```

```python
# find apple laptops that costs more than 1000
(lap['price']>1000)&(lap['Manufacturer']=='Apple')
```

```python
lap['Screen Size'].str.replace('"','')
```

```python
lap['Screen Size']=lap['Screen Size'].str.replace('"','')
```

```python
lap.head(3)
```

```python
# creating a conditional column from more than 2 choices : np.seleect()
```

```python
# create an array based on multiple price tiers (+2 choices)
```

```python
# create  'conditions ' and 'values' variables 

conditions=[
    lap['price']>2000,
    (lap['price']<=2000)&(lap['price']>1500),
    (lap['price']<=1500)&(lap['price']>1000),
    (lap['price']<=1000)
]
values=['Too Expensive','Expensive','Affordable','Cheap']
```

```python
# set it to a column
np.select(conditions,values)
```

```python
lap['report']=np.select(conditions,values)
```

```python
lap['price'].value_counts
```
