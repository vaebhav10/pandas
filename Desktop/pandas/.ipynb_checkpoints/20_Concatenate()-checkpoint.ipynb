```python
import pandas as pd
import numpy as np
```

1 Concat()

1.1 Concatenate vertically

```python
df1=pd.DataFrame({'id':['A','B','C','D'],
                 'value':[1,2,3,4]})
df2=pd.DataFrame({'id':['D','E','F'],
                 'value':[4,5,6]})
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)# axis=0 : vertical concatenation
# ignoreindex ignores the 
```

```python
pd.concat([df1,df2],axis=1)# concatenating horizontally 
```

1 Inner join

```python
df1.merge(df2,on='id',how='inner')
```

```python
df1.merge(df2,on='id',how='outer')# full join 
```

```python
df2.merge(df1,on='id',how='inner')
```

2 Full join(Exclusive)

```python
# indicator =true
df1.merge(df2,on='id',how='outer',indicator=True)
```

```python
# Query
df1.merge(df2,on='id',how='outer',indicator=True).query("_merge=='both'")
```

```python

df1.merge(df2,on='id',how='outer',indicator=True).query("_merge=='left_only'or _merge=='right_only'")
```

Left join

```python
df1.merge(df2,on='id',how='left',indicator=True)
```

Right join

```python
df1.merge(df2,on='id',how='right',indicator=True)
```

```python
#extract 50% from merged df1 and df2
merged=pd.concat([df1,df2],axis=0)
```

```python
merged.sample(frac=0.5)
```

Left join (Exclusive) {only in df1 }


```python
df1.merge(df2,on='id',how='outer',indicator=True)
```

```python
# data exclusive to df1
df1.merge(df2,on='id',how='outer',
          indicator=True).query ("_merge=='left_only'")
```
