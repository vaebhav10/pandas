```python
import numpy as np
import pandas as pd 
```

```python

lap=pd.read_csv("laptops.csv",encoding='latin1')
```

```python
lap=lap.rename(columns={'Price (Euros)':'price'})# renaming the price column
```

```python
lap['price']=lap['price'].str.replace(",",'.')
```

```python
lap['price'] =pd.to_numeric(lap['price'])
```

```python

# quesiton 
# find the cheapest and most expensive laptop of each company using sort_values and duplicated methods (keep "first " ,"last",and false )
```

```python
lap.sort_values(['Manufacturer', 'price'])#
```

```python
# check all values categories in 'Manufactrer' column

lap.value_counts('Manufacturer')
```

```python
# duplicated calues in manufacturer columns 
lap.duplicated('Manufacturer',keep='first')
```

```python
duplicate = lap.duplicated('Manufacturer',keep='first')
```

```python
lap[duplicate]
```

```python
# shows the cheapes laptop of Manufacturer
lap[~duplicate][['Manufacturer','price']]    # show dataframe with non duplicated values in "Manufacturer" column 
```

```python
high_dup=lap.duplicated('Manufacturer',keep='last')# now its the   price
```

```python
lap[high_dup]
```

```python
lap[~high_dup][['Manufacturer','price']]
```

```python
dupfalse=lap.duplicated('Manufacturer',keep=False)
```

```python
lap[~dupfalse]# since the keep is set to false
```

```python
# all elements in company column are repeated at least once 
lap.value_counts('Manufacturer')
```
