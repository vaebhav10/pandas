```python
import pandas as pd 
import numpy as np

df=pd.read_csv('file.csv')
del df['Unnamed: 0']
df=df.set_index('Student Name')
import random 
point1=[random.randint(150,180)for _ in range(13)]
df.rename(columns={'Points':'Marks'},inplace=True)
df['Height']=point1

df=df.dropna()# delete rows with any NaN values
```

```python
df.head(2)
```

1 query

```python
# select students with Marks greater than 55
df.query ('Marks>55')
```

```python
# the equivalent boolean slicing 
df[df['Marks']>55]
```

```python
# with multiple conditions
df.query('Marks>40 and Height>150')# or & 
```

```python
df[(df['Marks']>40)&(df['Height']>150)]
```

```python
df.query("not(Marks>60)") # with a not condition|
```

```python
#equivalent booleajn slicing of above 
df[~(df['Marks']>40)]
```

```python
# convert height to meters and select those with height above 1.6 meter
df.query('Height/100>1.6')
```

```python
# check the data types 
df.dtypes
# conver 'Date' column to date time type
df['Date'] =df['Date'].astype('datetime64[ns]')
df['Date'].dt.year
```

```python
df.query('Date.dt.year==2016')
```
