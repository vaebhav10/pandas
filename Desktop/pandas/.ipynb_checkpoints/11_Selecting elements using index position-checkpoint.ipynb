```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('file.csv')
del df['Unnamed: 0']
df=df.set_index('Student Name')
import random 
point1=[random.randint(150,180)for _ in range(13)]
df.rename(columns={'Points':'Marks'},inplace=True)
df['Height']=point1
```

2 Selecting with a list of values

```python
# get all data about Vaibhav Kumar and kelly frye

df.iloc [[0,2,4]]
```

```python
# get the height of vaibhav and kelly 
df.iloc[[0,2],3]
```

```python
# get the height and marks of vaibhav 
df.iloc[[0,2],[2,3]]
```

3 Selecting a range of data with a slice 

start:stop(start is included and stop is excluded )

```python
# slice column labels Marks/Height
# students =['Vaibhav Kumar','kelly frye']

df.iloc[[0,2],1:4]
```

```python
# same as above
std=[0,2]
df.iloc[std,2:4]
```

```python
# slice index labels (top1 to top5 students)
# columns=['Marks','Height']
col=[2,3]
df.iloc[0:6,col]
```

4 Selecting with conditions

iloc cannot accept a boolean Series , but only a boolean list .We have to 
use the list() function to convert a Series into a boolean list 

```python
# one condition : select student with height above 170cm

col=[2,3]
k=list(df['Height']>=170)
df.iloc[k,col]
```

```python
# multiple conditions : select player with height above 160cm and Marks >60
col=[1,2]


df.iloc[list((df['Marks']>50)&(df['Height']>150)),:]
```
