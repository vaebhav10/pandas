2. Selecting with a list of values

```python
# get all data about Vaibhav Kumar and kelly frye
```

```python
df.loc[['Vaibhav Kumar','kelly frye']]
```

```python
# get the marks of vaibhav and kelly
df.loc[['Vaibhav Kumar','kelly frye'],'Points']
```

```python
# get the points and date of vaibhav and kelly
df.loc[['Vaibhav Kumar','kelly frye'],['Points','Date']]
```

```python
# get the points and date of vaibhav kumar 
df.loc['Vaibhav Kumar',['Points','Date']]
```

3. Selecting a range of data with a slice 

```python
# start:stop:step(contrary to usual python slices , both the start and the stop are included )
```

```python
# slice column alabels
students=['Vaibhav Kumar','kelly frye']
# 'Date':'Points'
df.loc[students,'Date':'Points']
```

```python
# slice index lebels
```
