```python
import pandas as pd 
import numpy as np

df=pd.read_csv('file.csv')
del df['Unnamed: 0']
df=df.set_index('Student Name')
import random 
point1=[random.randint(150,180)for _ in range(13)]
df.rename(columns={'Points':'Marks'},inplace=True)
df['Height']=point1

df=df.dropna()# delete rows with any NaN values
```

```python
# create your own funcion and apply it to a df
# sum of all the marks 

def calculate_sum(row):
    return row['Marks']+row['Height'] 
```

```python
df.apply(calculate_sum,axis=1)
```
