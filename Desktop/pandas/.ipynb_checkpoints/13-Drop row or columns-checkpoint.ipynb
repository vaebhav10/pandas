```python
import pandas as pd 
import numpy as np

df=pd.read_csv('file.csv')
del df['Unnamed: 0']
df=df.set_index('Student Name')
import random 
point1=[random.randint(150,180)for _ in range(13)]
df.rename(columns={'Points':'Marks'},inplace=True)
df['Height']=point1

df=df.dropna()# delete rows with any NaN values
```

1 Drop Row

```python
df
```

```python
# drop one row
# 1 axis parameter 
df.drop('Kirsten Mcdowell',axis=0)# 0 for row and 1 for column
```

```python
# 2 index parameter
df.drop(index=['Kirsten Mcdowell'])# same as above 
```

```python
# drop two or more rows and update data (inplace =True)
df.drop(['Vaibhav Kumar','kelly frye'],axis=0)
```

2 Drop Column

```python
# drop one column
#1 axis parameter 
df.drop('Exam Name',axis=1)
# df.drop(columns=['Exam Name']) same as above
```

```python
# drop columns by position/index (drop last column)
# df.columns[[-1]]
df.drop(df.columns[[2]],axis=1)
```

```python
# drop two or more columns 
df.drop(['Exam Name','Height'],axis=1)
```
