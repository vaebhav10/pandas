```python
# creating a dataframe from an array 
```

```python
import numpy as np 
import pandas as pd
```

```python
data=np.array ([[1,2],[3,4],[6,7]])# creating an array
```

```python
# creating a dataframe
pd.DataFrame(data)
```

```python
df=pd.DataFrame(data,index=['row1','row2','row3'],
                columns=['col1','col2'])
```

```python
df
```

```python
data2=np.array([['name','vaibhav'],['middle_name','Kumar'],['surname','Ram']])
```

```python
d2=pd.DataFrame(data2,index=['R1','R2','R3'], columns=['I','II'])
# pd.DataFrame(data2)
```
