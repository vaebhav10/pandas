```python
# importing matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('population_total.csv')
df=df.dropna()
```

```python
# df_pivot=df.pivot[[df['country'].isin(['India','United States','Brazil','China'])]
```

```python
df_pivot=df.pivot(index='year',columns='country',values='population')
```

```python
df_pivot=df_pivot[['India','United States','Brazil','China']]
```

```python
df_pivot.plot(kind='line',
             xlabel='years',
             ylabel='Population',
             title='Population of countries',
             figsize=(8,4))
```

```python
df_pivot.plot(kind='line',
             xlabel='years',
             ylabel='Population',
             title='Population of countries',
             figsize=(8,4))
#save plot
plt.savefig('Saved graph.png')
# show plot
plt.show()
```
