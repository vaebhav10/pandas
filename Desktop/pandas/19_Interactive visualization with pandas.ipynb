```python
df=pd.read_csv('population_total.csv')
```

```python
# df=df.dropna()
df_pivot=df.pivot(index='year',columns='country',values='population')
df_pivot=df_pivot[['India','United States','Brazil','China']]
df_pivot
```

1 Lineplot

```python
df_pivot.iplot(kind='line',xTitle='Years',yTitle='Population')
```

2 Barplot 

```python
# df_pivot_20=
df_pivot_20=df_pivot[df_pivot['year'].index.isin([2020])]
```

```python
df_pivot_20=df_pivot_20.T
```

```python
df_pivot_20.iplot(kind='bar',color='lightgreen',xTitle='Year',yTitle='Population')
```
